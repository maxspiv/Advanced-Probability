```python
import random 
import numpy as np
import matplotlib.pyplot as plt
```

```python
def getUniformRandomNumber():
    rand = random.uniform(0,1)
    return rand
```

```python
def getSampleMean(N):
    #arr = np.random.uniform(0,1,N)
    #arr = np.random.exponential(1,N)
    #arr = np.random.standard_cauchy(N)
    #arr = np.random.pareto(1,N)
    arr = np.random.zipf(10,N)
    return arr.mean()
```

```python
getSampleMean(100)
```

```python
def histogram(N):
    a = []
    for i in range(10000):
        a.append(getSampleMean(N))
    plt.hist(a,density=1,bins=10)
    plt.show
```

```python
histogram(100)
```
