## For every continuous probability distribution is scipy.stats that can be constructed with default parameters I calculated and plotted probabilities of being above multiples of standard deviation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats
```

```python
def plot_dist(dist_name:str,stdv:np.array):
    out = []
    dist = getattr(scipy.stats,dist_name)
    std_val = dist.std()
    mean_val = dist.mean()
    for stdvi in stdv:
        out.append(dist.cdf(-stdvi*std_val+mean_val)+dist.sf(stdvi*std_val+mean_val))
    plt.plot(stdv,np.log(np.array(out)))
    plt.xlabel('k')
    plt.ylabel(r'$\log ( P(|X-\mu| > k \cdot \sigma)  )$')
    plt.title(dist_name)
```

```python
k_vals = np.arange(1,10)
```

```python
plot_dist('expon',k_vals)
```
