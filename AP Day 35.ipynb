```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats
```

```python
def func(dist,stdv):
    out = []
    dist = scipy.stats.norm #just for now
    for i in range(len(stdv)):
        out.append(1-(dist.cdf(stdv[i]*dist.std())-dist.cdf(-stdv[i]*dist.std())))
    return stdv, out
```

```python
plot = func('test',[1,2,3,4,5])
```
