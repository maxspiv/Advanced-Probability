```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats
```

```python
def func(dist,stdv):
    out = []
    dist_continu = [d for d in dir(scipy.stats) if isinstance(getattr(scipy.stats, d), scipy.stats.rv_continuous)]
    distr_name = dist
    dist = getattr(scipy.stats,distr_name)
    for i in range(len(stdv)):
        out.append(1-(dist.cdf(stdv[i]*dist.std())-dist.sf(stdv[i]*dist.std())))
    return stdv, out
```

```python
plot = func('norm',[1,2,3,4,5])
```
